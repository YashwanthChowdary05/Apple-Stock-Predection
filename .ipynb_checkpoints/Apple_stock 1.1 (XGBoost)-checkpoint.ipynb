# **Handle Missing Values (Blanks)**

```python
#counting missing values
print("Missing values per column:")
print(df.isnull().sum())
```

Market sentiment column can fluctuate  use mean imputation

```python
#checking exactly which row has missing values
df[df.isna().any(axis=1)]
```

### 1. Filling missing values Time-based interpolation for stock and market
for columns stock_price , nasdaq_index , sp500_index.
Fills missing values by estimating what the value should have been at that timestamp, based on nearby known values.

```python
# Filling stock_price columns logically
df['stock_price'] = df['stock_price'].interpolate(method='time')
```

```python
#checking the stock_price column is filled or not
df.loc[['2010-01-07 10:00:00','2010-01-11 10:00:00']]
```

Here the stock_price are successfully filled by interpolation

```python
#similarly for nasdaq_index and sp500_index
df['nasdaq_index'] = df['nasdaq_index'].interpolate(method='time')
df['sp500_index'] = df['sp500_index'].interpolate(method='time')
```

### 2. Forward fill for slow-moving economic indicators
for columns inflation_rate , unemployement_rate , interest_rate.
Forward fill (ffill) replaces missing values with the last known valid value.

```python
df['inflation_rate'] = df['inflation_rate'].fillna(method='ffill')
df['unemployment_rate'] = df['unemployment_rate'].fillna(method='ffill')
df['interest_rate'] = df['interest_rate'].fillna(method='ffill')
```

### 3. Mean imputation for market sentiment

for market_sentiment column
Replaces missing sentiment values with the average sentiment score across the dataset.

```python
df['market_sentiment'] = df['market_sentiment'].fillna(df['market_sentiment'].mean())
```

# **Handle Outliers (Don’t Drop)**

```python
#generating boxplots for all numeric columns
import matplotlib.pyplot as plt
import seaborn as sns

# List of numeric columns
numeric_cols = ['stock_price', 'nasdaq_index', 'sp500_index',
                'inflation_rate', 'unemployment_rate',
                'interest_rate', 'market_sentiment']

# Set figure size
plt.figure(figsize=(15, 8))

# Create boxplots for each numeric column
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 4, i)  # 2 rows, 4 columns layout
    sns.boxplot(y=df[col])
    plt.title(col)
    plt.tight_layout()

plt.show()
```

From the boxplots, it is evident that outliers are present only in the stock_price column. Therefore, we will use the IQR method to handle and cap these outliers appropriately.

```python
# Calculate Q1, Q3, and IQR for 'stock_price'
Q1 = df['stock_price'].quantile(0.25)
Q3 = df['stock_price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
df['stock_price'] = df['stock_price'].clip(lower=lower_bound, upper=upper_bound)

plt.figure(figsize=(5, 5))
sns.boxplot(y=df['stock_price'])
plt.title('Stock Price After Outlier Capping')
plt.show()
```

# **Exploratory Data Anlaysis**

```python
#trend, sesaonality , residual for stock_price column
#Assuming hourly data, we can try period = 24 (daily seasonality)
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['stock_price'], model='additive', period=24)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plotting
plt.figure(figsize=(15,12))

plt.subplot(4,1,1)
plt.plot(df['stock_price'], label='Original', color='blue')
plt.title('Original Stock Price')
plt.legend()

plt.subplot(4,1,2)
plt.plot(trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.legend()

plt.subplot(4,1,3)
plt.plot(seasonal, label='Seasonality', color='green')
plt.title('Seasonal Component')
plt.legend()

plt.subplot(4,1,4)
plt.plot(residual, label='Residual', color='red')
plt.title('Residual Component')
plt.legend()

plt.tight_layout()
plt.show()
```

Period is set to 24 for hourly data with daily seasonality.

If you want weekly seasonality, set period=24*5 (5 trading days).

For monthly trends, use period=24* 5 *4 (approx 4 weeks).

here the dataset follows the Upward Linear Trend and non-seasonal

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error # Includes MAE
import matplotlib.pyplot as plt

# --- 1. Data Loading and Preprocessing (Replicated from notebook) ---
try:
    # IMPORTANT: Ensure the file 'Apples_stock price dataset.xlsx' is in your directory
    df = pd.read_excel("Apples_stock price dataset.xlsx", sheet_name="Sheet1")
except FileNotFoundError:
    raise # Will stop execution if file is missing

df.columns = [c.strip().lower() for c in df.columns]
df.set_index(pd.to_datetime(df['timestamp'], errors='coerce'), inplace=True)
df.drop(columns=['timestamp'], inplace=True, errors='ignore')
df = df.between_time('04:00', '20:00')
df = df[df.index.dayofweek < 5] # Exclude weekends

# Handle Missing Values
df['stock_price'] = df['stock_price'].interpolate(method='time')
df['nasdaq_index'] = df['nasdaq_index'].interpolate(method='time')
df['sp500_index'] = df['sp500_index'].interpolate(method='time')
df['inflation_rate'] = df['inflation_rate'].ffill()
df['unemployment_rate'] = df['unemployment_rate'].ffill()
df['interest_rate'] = df['interest_rate'].ffill()
df['market_sentiment'] = df['market_sentiment'].fillna(df['market_sentiment'].mean())

# Handle Outliers (Capping for 'stock_price')
Q1 = df['stock_price'].quantile(0.25)
Q3 = df['stock_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['stock_price'] = df['stock_price'].clip(lower=lower_bound, upper=upper_bound)

# --- 2. Feature Engineering Functions ---
def create_time_features(df):
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df

def create_lag_and_rolling_features(df, target_col):
    trading_hours_per_day = 17 
    # Lag features
    df['stock_price_lag24'] = df[target_col].shift(trading_hours_per_day)
    df['stock_price_lag48'] = df[target_col].shift(2 * trading_hours_per_day)
    df['nasdaq_lag24'] = df['nasdaq_index'].shift(trading_hours_per_day)
    df['sp500_lag24'] = df['sp500_index'].shift(trading_hours_per_day)
    # Rolling mean
    df['stock_price_rolling_mean_5d'] = df[target_col].shift(1).rolling(window=5*trading_hours_per_day).mean()
    return df

# Apply feature engineering
df_model = create_time_features(df.copy())
df_model = create_lag_and_rolling_features(df_model, 'stock_price')
df_model = df_model.dropna()

FEATURES = [
    'nasdaq_index', 'sp500_index', 'inflation_rate', 'unemployment_rate',
    'interest_rate', 'market_sentiment', 
    'hour', 'dayofweek', 'quarter', 'month', 'year', 'dayofyear',
    'stock_price_lag24', 'stock_price_lag48', 'nasdaq_lag24', 'sp500_lag24',
    'stock_price_rolling_mean_5d'
]
TARGET = 'stock_price'
X = df_model[FEATURES]
y = df_model[TARGET]

# --- 3. Train-Test Split (Chronological Split for Metrics) ---
split_point = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

# --- 4. Model Training for Evaluation ---
gbr_eval = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, subsample=0.7, random_state=42)
gbr_eval.fit(X_train, y_train)

# --- 5. Evaluate Model Performance ---
y_pred_test = gbr_eval.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
mae = mean_absolute_error(y_test, y_pred_test)

# --- 6. Train FINAL Model on FULL Data for Best Forecast ---
gbr_final = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, subsample=0.7, random_state=42)
gbr_final.fit(X, y)

# --- 7. Future Prediction for Next 30 Days (Recursive Forecasting) ---
trading_hours_per_day = 17
hours_to_predict = 30 * trading_hours_per_day 
last_timestamp = df.index[-1]
future_timestamps = pd.date_range(start=last_timestamp + pd.Timedelta(hours=1), periods=hours_to_predict * 2, freq='H')
future_timestamps = future_timestamps[(future_timestamps.hour.isin(range(4, 21))) & (future_timestamps.dayofweek < 5)][:hours_to_predict]

X_future = pd.DataFrame(index=future_timestamps)
X_future['nasdaq_index'] = df_model['nasdaq_index'].iloc[-1]
X_future['sp500_index'] = df_model['sp500_index'].iloc[-1]
mean_exog = df_model[['inflation_rate', 'unemployment_rate', 'interest_rate', 'market_sentiment']].iloc[-50:].mean()
for col in mean_exog.index:
    X_future[col] = mean_exog[col]
X_future = create_time_features(X_future.copy())

current_df = df_model.copy()
forecasted_values = []

# --- Recursive Forecasting Loop ---
for timestamp in X_future.index:
    X_step_data = X_future.loc[timestamp, ['nasdaq_index', 'sp500_index', 'inflation_rate', 'unemployment_rate',
                                          'interest_rate', 'market_sentiment', 'hour', 'dayofweek', 'quarter', 
                                          'month', 'year', 'dayofyear']]
    X_step = pd.DataFrame([X_step_data.values], columns=X_step_data.index, index=[timestamp])
    
    # Calculate and add LAG/ROLLING features from the current_df
    X_step['stock_price_lag24'] = current_df['stock_price'].iloc[-trading_hours_per_day]
    X_step['stock_price_lag48'] = current_df['stock_price'].iloc[-2*trading_hours_per_day]
    X_step['nasdaq_lag24'] = current_df['nasdaq_index'].iloc[-trading_hours_per_day]
    X_step['sp500_lag24'] = current_df['sp500_index'].iloc[-trading_hours_per_day]
    window_size = 5 * trading_hours_per_day
    X_step['stock_price_rolling_mean_5d'] = current_df['stock_price'].iloc[-window_size:].mean()

    # Predict using the final model trained on all data
    predicted_price = gbr_final.predict(X_step[FEATURES])[0]
    forecasted_values.append(predicted_price)
    
    # Prepare the new row to append to current_df
    new_row_data = current_df.iloc[-1].to_dict()
    new_row_data[TARGET] = predicted_price
    new_row_data['nasdaq_index'] = X_step['nasdaq_index'].iloc[0]
    new_row_data['sp500_index'] = X_step['sp500_index'].iloc[0]
    
    new_row = pd.Series(new_row_data, name=timestamp)
    current_df = pd.concat([current_df.iloc[1:], pd.DataFrame(new_row).T])

forecast_series = pd.Series(forecasted_values, index=X_future.index)

# --- 8. Plotting the Results and Printing Metrics ---
plt.figure(figsize=(15, 6))
last_30_days_count = trading_hours_per_day * 30
last_30_days = df['stock_price'].tail(last_30_days_count) 
plt.plot(last_30_days, label='Historical Stock Price (Last 30 trading days)', color='blue')
plt.plot(forecast_series, label='Gradient Boosting 30-Day Forecast', color='red', linestyle='--')
plt.title('Apple Stock Price: Last 30 Trading Days Actuals and Next 30 Trading Days Forecast')
plt.xlabel('Timestamp')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
print(f"Model Performance Metrics on 20% Test Data:")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")
print(f"MAE (Mean Absolute Error): {mae:.2f}")

mean = df['stock_price'].mean()

mape = (mae/mean)*100
rmspe = (rmse/mean)*100

print(f'Mean Absolute Percentage Error (MAPE): {mape: 2f}%')
print(f'Root Mean Squared Percentage Error (RMSPE): {rmspe: 2f}%')
```
